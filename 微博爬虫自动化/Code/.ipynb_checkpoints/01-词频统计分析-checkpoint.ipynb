```python
import jieba
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from docx import Document
import pandas as pd
import random
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
plt.style.use('ggplot')
```

## 预处理

```python
### 读取文档
document = Document("./中共中央关于党的百年奋斗重大成就和历史经验的决议(1).docx")
```

```python
all_paragraphs = document.paragraphs
```

```python
text_list = []
for paragraph in all_paragraphs:
    for run in paragraph.runs:
        text_list.append(run.text)
```

```python
text_list = [str(i).strip().replace('\n','').replace('\t','') for i in text_list]
```

```python
with open('中共中央关于党的百年奋斗重大成就和历史经验的决议.txt','w+',encoding='utf-8') as fp:
    fp.write(str(text_list))
```

```python
# 创建停用词列表
def get_stopwords_list():
    stopwords = [line.strip() for line in open('stopwords.txt',encoding='UTF-8').readlines()]
    return stopwords
stopwords = get_stopwords_list()
```

```python
stopwords[:10]
```

```python
def seg_depart(sentence):
    # 对文档中的每一行进行中文分词
    sentence_depart = jieba.lcut(sentence)
    return sentence_depart
```

```python
text_list2=[]
# text_list3 = []
for text in text_list:
    text = seg_depart(text)
    for index,one in enumerate(text):
        if one not in stopwords:
            item = {
                '序号':index,
                '词语':one
            }
#             text_list3.append(one)
            text_list2.append(item)
```

```python
df =pd.DataFrame(text_list2)
```

```python
row_df = df.groupby('词语').count().sort_values('序号',ascending=False)
row_df.reset_index(inplace=True)
row_df.columns = ['词语','次数']
```

## 词频最高的20个词语

```python
row_df_TOP_20 = row_df.head(20)
```

```python
value1 = list(row_df_TOP_20['词语'].values)
value1
```

## TOP20柱状图

```python
row_df_TOP_20.head(3)
```

```python
plt.figure(figsize=(15,10))
plt.title('词频TOP20柱状图')
plt.bar(list(row_df_TOP_20['词语'].values),list(row_df_TOP_20['次数'].values))
plt.ylabel('词频')
plt.xlabel('词语')
plt.show()
```

## 随机抽选可视化

```python
index = list(row_df.index.values)
```

```python
random_index = random.choices(index,k=10)
word_list_r = []
num_list_r = []
for one_index in random_index:
    word = row_df.iloc[one_index]['词语']
    word_list_r.append(word)
    num = row_df.iloc[one_index]['次数']
    num_list_r.append(num)
plt.figure(figsize=(10,6))
plt.bar(word_list_r,num_list_r)
plt.title('随机抽样词频图')
plt.ylabel('词频')
plt.xlabel('词语')
```
