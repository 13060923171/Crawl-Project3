```python
import requests 
import bs4 as bs 

from time import sleep
from time import time
from random import randint

import pandas as pd 
```

```python
titles = []
durations = []
playtimes = []
subtitles = []
dates = []


pages = [str(i) for i in range(1,28)] 
headers = {"Accept-Language": "en-US, en;q=0.5"}
for page in pages:
    source = requests.get('https://search.bilibili.com/all?keyword=%E9%A6%99%E6%B8%AF%E5%9F%8E%E5%B8%82%E5%A4%A7%E5%AD%A6&page=' + page, headers = headers)
    sleep(randint(1,3))
    page_html = bs.BeautifulSoup(source.content, 'html.parser')
    news_containers = page_html.find_all('ul', class_="video-list clearfix")
    for container in news_containers:
        for row in container.find_all('li'):
            title = row.select('a:nth-of-type(1)')[1].text
            #print("date")
            print(title)
            titles.append(title)
            duration = row.select('a:nth-of-type(1)')[0].text
            print(duration)
            durations.append(duration)
        for row in container.find_all("span", class_="so-icon watch-num"):
            print(row.text.strip())
            playtimes.append(row.text.strip())
        for row in container.find_all("span", class_="so-icon hide"):
            print(row.text.strip())
            subtitles.append(row.text.strip())
        for row in container.find_all("span", class_="so-icon time"):
            print(row.text.strip())
            dates.append(row.text.strip())  
       
        
```

```python
# 1.8万 转换为 18000
str = '1.8万'
if str.__contains__('万'):
    num_str = str[:-1]
    num = int(float(num_str) * 10000)
    
```

```python
print(len(titles))
print(len(durations))
print(len(playtimes))
print(len(subtitles))
print(len(dates))
```

```python
# create a dictionary
databz = {
    'TITLES': titles,
    '时长': durations,
    '播放量': playtimes,
    '弹幕数': subtitles,
    '发布日期': dates
    }


newTitles = []
newDurations = []
newPlaytimes = []
newSubtitles = []
newDates = []
```

```python
for i in range(len(titles)):
    # 城大香港城市大学CityU香港城市大學
    if titles[i].__contains__('城大') \
            | titles[i].__contains__('香港城市大学') \
            | titles[i].__contains__('CityU') \
            | titles[i].__contains__('香港城市大學'):
        newTitles.append(titles[i])
        newDurations.append(durations[i])
        newPlaytimes.append(playtimes[i])
        newSubtitles.append(subtitles[i])
        newDates.append(dates[i])
```

```python
databz = {
    'TITLES': newTitles,
    '时长': newDurations,
    '播放量': newPlaytimes,
    '弹幕数': newSubtitles,
    '发布日期': newDates
}
```

```python
import pandas as pd 
pd_CityU = pd.DataFrame.from_dict(databz)
```
