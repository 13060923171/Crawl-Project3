```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#图表窗口
plt.plot(np.random.rand(10))
plt.show()
```

```python
#图表窗口2 -魔法函数，嵌入图表
%matplotlib inline
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)
#直接嵌入图表，不用plt.show()
```

```python
#图表窗口3 - 魔法函数，弹出可以交互的matplotlib窗口
%matplotlib notebook
s = pd.Series(np.random.randn(100))
s.plot(style = 'k--o',figsize=(10,5))
#可交互的matplotlib窗口，不用plt.show()
#可做一定调整
```

```python
#图表窗口4 -魔法函数，弹出matplotlib控制台
%matplotlib qt5
df = pd.DataFrame(np.random.rand(50,2),columns=['A','B'])
df.hist(figsize=(12,5),color = 'g',alpha=0.8)
#可交互性控制台
#如果已经设置了显示方式（比如notebook）,需要重新启动然后再运行魔法函数
#网页嵌入的交互性窗口和控制台，只能显示一个
```
